```python
import os
import glob
from collections import Counter
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
```

```python
#dataframe pontuacao SEM RETICENCIAS

pontuacao_final = pd.DataFrame()

for file in os.scandir("./TXT-punct"):
    if file.is_file():
        with open(file) as fp:
            content = fp.read()
            content_list = list(content)
            dict_lst = Counter(content_list)

            full_data = pd.DataFrame.from_dict(dict_lst, orient='index').reset_index()
            full_data['filename'] = os.path.basename(file)
            full_data = full_data.assign(filename_lista=full_data.filename.str.split('-'))
            full_data[['ano', 'titulo', 'autor', 'punct']] = pd.DataFrame(full_data.filename_lista.tolist(), index= full_data.index)
            full_data.rename(columns={'index': 'pontuacao', 0: 'ocorrencias'}, inplace=True)
            
            #acertos e filtragens
            full_data["."] = (full_data["pontuacao"] == '.')
            full_data = full_data.loc[(full_data["pontuacao"] == ',') | (full_data["pontuacao"] == '.') | (full_data["pontuacao"] == ':') | (full_data["pontuacao"] == ';') | (full_data["pontuacao"] == '?') | (full_data["pontuacao"] == '!') | (full_data["pontuacao"] == '(') | (full_data["pontuacao"] == ')')]
            full_data = full_data.loc[((full_data["ocorrencias"] > 5))]
            
            full_data = full_data[['ano', 'titulo', 'autor','pontuacao','ocorrencias']]

            pontuacao_final = pd.concat([pontuacao_final, full_data])
            


pontuacao_final = pd.pivot_table(pontuacao_final, values="ocorrencias", index=["ano", "titulo", "autor"], columns=["pontuacao"])
#pontuacao_final.to_csv("pontuacao_final_SEM_RETICENCIAS.csv", index=False)
pontuacao_final
```

```python
#contagem de palavras e afins

textos_limpos = pd.DataFrame()

#ir buscar os ficheiros TXT
file_location = os.path.join('TXTs', '*.txt')


#esta funçao leva alguns segundos a correr
for file in glob.glob(file_location):
    with open(file) as fp:
            
            textinhos = fp.read()
            textinhos_dict = {}

            #limpar o texto para dataframe
            textinhos_dict["reticencias"] = textinhos.count("...")
            textinhos_dict["texto"] = re.sub(r'[^\w\s]', '', textinhos).lower()
            textinhos_dict["texto"] = re.sub(r'[\n]', ' ', textinhos).lower()
            textinhos = pd.DataFrame.from_dict(textinhos_dict, orient='index')
            textinhos = textinhos.transpose()

            # ano, título e autor
            textinhos['filename'] = os.path.basename(file)
            textinhos = textinhos.assign(filename_lista=textinhos.filename.str.split('-'))            
            textinhos[['ano', 'titulo', 'autor']] = pd.DataFrame(textinhos.filename_lista.tolist(), index= textinhos.index)
            textinhos["autor"] = textinhos.autor.str.strip(".txt")
            textinhos.rename(columns={0: 'texto'}, inplace=True)

            # número de palavras e palavras únicas
            
            textinhos["palavras_num"] = [len(x) for x in textinhos["texto"]]

            stopwords_pt = nltk.corpus.stopwords.words('portuguese')
            stopwords_pt.append("porque")
            textinhos['no_stopwords'] = textinhos.apply(lambda x : ' '.join([word for word in x['texto'].split(' ') if word not in stopwords_pt]), axis = 1)
            textinhos['tokens'] = textinhos['no_stopwords'].apply(nltk.word_tokenize)

            textinhos["palavras_unic"] = [len(x) for x in textinhos['tokens']]

            
            textinhos = textinhos[['ano', 'titulo', 'autor', 'reticencias', 'palavras_num', 'palavras_unic']]
           
            textos_limpos = pd.concat([textos_limpos, textinhos])

textos_limpos
```

```python
#conjunto de dados final limpinho

new_df = pd.merge(pontuacao_final, textos_limpos, on=['ano', 'titulo', 'autor'], how="left", sort=False)
new_df = new_df.fillna(0)
new_df["."] = new_df["."] - new_df["reticencias"]
new_df = new_df.astype({'ano': 'int', '!': 'int', '(': 'int', ')': 'int', ',': 'int', '.': 'int', ':': 'int', ';': 'int', '?': 'int'})

new_df["pont_total"] = new_df["!"] + new_df["("] + new_df[")"] + new_df[","] + new_df["."] + new_df[":"] + new_df[";"] + new_df["?"]
new_df["pont_frases"] = new_df["!"] +  new_df["."] + new_df["?"] + new_df["reticencias"]
new_df["palavras_por_pont"] = new_df["palavras_num"] / new_df["pont_total"]
new_df["palavras_por_frase"] = new_df["palavras_num"] / new_df["pont_frases"]
#new_df["racio_unic"] = new_df["pont_total"] / new_df["palavras_unic"]
```

```python
palavras_resumo = new_df[["ano", "titulo", "autor", "palavras_num", "palavras_unic", "pont_total", "pont_frases", "palavras_por_pont","palavras_por_frase"]]
palavras_resumo.to_csv("palavras_resumo.csv", index=False)
```
